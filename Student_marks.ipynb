```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as stats

#DATASET
marks=[25,40,48,50,51,55,55,65,67,69,72,75,78,78,80,81,85,87,91,95]
mean_val=stats.mean(marks)
medium_val=stats.median(marks)
mode_val=stats.mode(marks)
range_val=max(marks)-min(marks)
sts_val=stats.stdev(marks)
variance_val=stats.variance(marks)

#STORE THE RESULTS
result={
    "Mean":mean_val,
    "Median":medium_val,
    "Mode":mode_val,
    "Range":range_val,
    "Standard Deviation":sts_val,
    "Variance":variance_val
}

#PLOT THE HISTOGRAM
plt.hist(marks,bins=5,color='green',edgecolor='black')
plt.xlabel('Marks')
plt.ylabel('frequency')
plt.title('Histogram of Students Marks')
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.show()

print(result)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as stats

#DATASET
marks=[25,40,48,50,51,55,55,65,67,69,72,75,78,78,80,81,85,87,91,95]
mean_val=stats.mean(marks)
medium_val=stats.median(marks)
mode_val=stats.mode(marks)
range_val=max(marks)-min(marks)
sts_val=stats.stdev(marks)
variance_val=stats.variance(marks)

#STORE THE RESULTS
result={
    "Mean":mean_val,
    "Median":medium_val,
    "Mode":mode_val,
    "Range":range_val,
    "Standard Deviation":sts_val,
    "Variance":variance_val
}

#PLOT THE HISTOGRAM
plt.hist(marks,bins=5,color='purple',edgecolor='black')
plt.xlabel('Marks')
plt.ylabel('frequency')
plt.title('Histogram of Students Marks')
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.show()

print(result)

#BOX PLOT
plt.boxplot(marks,vert=True)
plt.xlabel('Marks')
plt.title('Box plot of students marks')
plt.grid(axis='x',linestyle='--',alpha=0.7)
plt.show()
```
