```python
#1. student Grade

name = input("Enter student's name")
marks = float(input("Enter student's percentages"))

if marks >= 90:
    grade = "A+"
elif marks >= 80:
    grade = "A"
elif marks >= 70:
    grade = "B"
elif marks >= 60:
    grade = "C"
elif marks >= 50:
    grade = "D"
else:
    grade = "Fail"

print(f"\nStudent: {name}")
print(f"Marks: {marks}%")
print(f"Grade: {grade}")
```

```python
#2. Simple ATM Simulator

balance = 1000.0 # starting balance

print("Welcome to the ATM!")
print("1. Check Balance")
print("2. Deposit Money")
print("3. Withdraw Money")
print("4. Exit")

choice = int(input("choose an option (1-4): "))

if choice == 1:
  print(f"Your balance is R{balance:.2f}")
elif choice == 2:
  deposit = float(input("Enter amount to deposit"))
  balance += deposit
  print(f"Deposited R{deposit:.2f} . New balance : R{balance:.2f}")
elif choice == 3:
  Withdraw = float(input("Enter amount to withdraw : "))
  if Withdraw <= balance :
    balance -= Withdraw
    print(f"Withdrew R{Withdraw:.2f} . Remaining balance : R {balance:.2f}")
  else:
    print("Insufficient funds!")
elif choice == 4:
    print("Thank you for using MYBANK ATM.")
else:
  print("Invalid option selection .")
```

```python
#3. Traffic Light Simulator

color = input("Enter traffic light color(red , yellow , green): ").lower()

if color == "red" :
  print("STOP ! Wait until the light changes.")
elif color == "yellow" :
  print("Prepare to STOP - light will turn red soon.")
elif color == "green" :
  print("Go! you may proceed carefully ")
else:
  print ("Invalid color! please enter red , yellow , or green .")
```

```python
# 4. Restuarant Order System

print("Welcome to Pizza Palace")
print("Menu")
print("1. Magherita - R60")
print("2. Chicken BBQ - R85")
print("3. Pepperoni - R90")

choice = int(input("Choose your Pizza(1-3): "))

if choice == 1:
  cost = 60
  pizza = "Magherita"
elif choice == 2:
  cost = 85
  pizza = "Chicken BBQ"
elif choice == 3:
  cost = 90
  pizza = "Pepperoni"
else:
  print("Invalid choice")
  cost = 0
  pizza = None

if pizza:
  extra_cheese = input("Add extra cheese for R10? (YES/NO): ").lower()
  if extra_cheese == "yes":
    cost += 10
  print(f"\nOrder Summury:\nPizza: {pizza}\nTotal Cost: R{cost}")
```
